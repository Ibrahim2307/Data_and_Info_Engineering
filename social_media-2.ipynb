```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
df = pd.read_csv('sentimentdataset.csv')

# Define a function to categorize month into seasons
def get_season(month):
    if month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Autumn'
    else:
        return 'Winter'

# Apply the function to create a new column for seasons
df['Season'] = df['Month'].apply(get_season)

# Count the number of posts in each season
season_counts = df['Season'].value_counts()

# Plotting the distribution as a pie chart
plt.figure(figsize=(8, 6))
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Posts by Season')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data from a CSV file
df = pd.read_csv('sentimentdataset.csv')

# Group the data by platform and calculate the sum of likes
data = df.groupby('Platform')['Likes'].sum()

# Create a pie chart
plt.pie(data, labels=data.index, autopct='%1.1f%%')
plt.title('Distribution of Likes by Platform')
plt.show()
```
